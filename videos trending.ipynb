# Which year was data Collected

```python
cdf = trending_videos_report['trending_date'].apply(lambda x: "20"+x[:2]).value_counts().to_frame().reset_index().rename(columns = {"index":'year' , "trending_date":"no_of_videos"})
```

```python
cdf
```

```python
x = cdf['year']
y = cdf['no_of_videos']

plt.bar(x,y)

plt.xlabel("Years")
plt.ylabel("no of videos")

plt.show()
```
